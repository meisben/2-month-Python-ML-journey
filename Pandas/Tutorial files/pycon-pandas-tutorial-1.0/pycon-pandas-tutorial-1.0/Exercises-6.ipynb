### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
s1 = sales1.rename(columns={
    "Book title": "title",
    "Number sold": "number",
    "Sales price": "their_price",
    "Royalty paid": "we_got"
})
s1['currency'] = 'USD'
s1
```

```python
s2 = sales2.copy()
t = sales2.Title
t
```

```python
t = t.where(t.str.endswith(')')).str.split().str[-1].str.strip('()')
t
```

```python
s2['currency'] = t.fillna(method='bfill')
s2
```

```python
s2 = s2[s2['List price'].notnull()]
s2
```

```python
s2 = s2.rename(columns={
    'Units sold': 'number',
    'Title': 'title',
    'List price': 'their_price',
    'Royalty': 'we_got',
    })
s2
```

```python
df = pd.concat([s1, s2])
df
```

```python
df['total_paid'] = df.their_price * df.number
df
```

```python
df.groupby(['title','currency']).sum()
```

```python
df.groupby(['title','currency'])[['total_paid']].sum()
```
